```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
df=pd.read_csv(url, header=None)
pd.set_option('display.max_columns', 500)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
df.head(5)
```

df.replace(x,y), remplaza datos x por y  

```python
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
missing_data=df.isnull()
missing_data.head()
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)  #remplazamos las filas que no tienen perdidas normalizadas por la media de la columna
df["normalized-losses"].replace(np.nan, round(avg_norm_loss), inplace=True)
df.head()
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
avg_bore
```

```python
df.bore.replace(np.nan, round(avg_bore,2), inplace=True)
df.head()
```

```python
stroke_mean=df['stroke'].astype('float').mean()
df.stroke.replace(np.nan,round(stroke_mean,2),inplace=True)
df.head()
```

```python
avg_horsepower = df.horsepower.astype('float').mean(axis=0)
avg_horsepower
```

```python
df['horsepower'].replace(np.nan, round(avg_horsepower), inplace=True)
df.head()
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
avg_peakrpm
```

```python
df['peak-rpm'].replace(np.nan, round(avg_peakrpm), inplace=True)
df.shape
```

```python
df["num-of-doors"].replace(np.nan, df['num-of-doors'].value_counts().idxmax(), inplace=True)   #Calcula el numero mas frecuente de puertas y lo remplaza en donde hay nan
```

```python
df.dropna(subset=["price"], inplace=True)   #borra las filas que no tienen precio
df.reset_index(drop=True, inplace=True)     #Resetea el indice por las filas que se borraron
df.shape
```

```python
df.dtypes
```

Corregimos tipos de variables

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df['city-L/100km'] = 235/df["city-mpg"]      # agregar columana con conversion a unidades metricas
df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]      #cambia unidades a metricas y renombra columna
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

```python
df['length'] = df['length']/df['length'].max()          #normaliza los valores 
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df.head()
```

```python
df["horsepower"]=df["horsepower"].astype(int)
df["horsepower"].value_counts()
```

```python
plt.pyplot.hist(df["horsepower"])
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
pyplot.bar(group_names, df["horsepower-binned"].value_counts())
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
plt.pyplot.hist(df["horsepower"], bins = 3)

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)     #junta la tabla dummy con df
df.drop("fuel-type", axis = 1, inplace=True)        
df.head()
```

```python
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)
df.drop("aspiration", axis = 1, inplace=True)
dummy_variable_2.rename(columns={'std':'std-type', 'turbo':'turbo-type'}, inplace=True)
dummy_variable_2.head()
```

```python
df.to_csv('clean_df.csv')
```
